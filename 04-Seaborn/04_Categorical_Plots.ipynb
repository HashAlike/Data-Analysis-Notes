A)Bar Plot

```python
sns.barplot(x='sex', y='total_bill',data=tips_df,estimator=np.cov)


# This is the most critical part of the graph. Normally, barplots take the mean of the data by default.
# The estimator parameter is the answer to the question, “Which mathematical function should I use when calculating the column height?”
```

E)Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x="day",y="total_bill" , data=tips_df, hue="sex", jitter=True , dodge=True)
```
