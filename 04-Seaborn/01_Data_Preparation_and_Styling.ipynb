B)Styling

```python
sns.set_style("white") 

plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )
```

```python
sns.set_style("dark") 

plt.figure(figsize=(8,4))

sns.set_context('talk',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )

sns.despine(left=True,bottom=False)
```

```python
sns.set_style("darkgrid") 

plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )

sns.despine(left=False,bottom=False)
```

```python
sns.set_style("whitegrid") 

plt.figure(figsize=(8,4))

sns.set_context('notebook',font_scale=1.4)


sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )

sns.despine(left=True,bottom=True)
```

```python
sns.set_style("ticks") 

plt.figure(figsize=(8,4))

sns.set_context('poster',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )

sns.despine(left=False,bottom=True)
```

sns.set_style(" ") -> Sets the background theme of the graph.

                   -> Possible values for x: darkgrid (default), whitegrid, dark, white, ticks.

                   -> Example: If you want grid lines, whitegrid is ideal.

plt.figure(figsize=())-> Specifies the physical size of the graph in inches.

                      -> Possible values for ( _ ): numbers such as (12, 5), (8, 6). The first number represents the width, the second number
                        represents the height.

sns.set_context('',font_scale=1.0)-> Adjusts the scale of graphic elements (lines, text).

                                  -> Possible replacements for x: paper, notebook, talk, poster.

                                  -> font_scale only determines the scaling factor for text.

sns.jointplot(x='speeding',y='alcohol', data= crash_df, kind="reg" )-> It displays the relationship between two variables as both a scatter plot and a distribution.

                                                                    -> Alternatives for kind: scatter (scatter plot), reg (regression line),
                                                                       resid (residuals), kde density), hex (hexagonal honeycomb).

sns.despine(left=F/T,bottom=F/T, right=F/T, top=F/T)-> Removes the outer frame lines (axes) of the graph.

                                                    -> Parameters: If you enter True (T), that edge is deleted; if you enter False (F),
                                                       that edge remains. It is generally used to clear the top and right lines.
