#### Statictics

```python
stats= np.array([[1,2,3],[4,5,6]])
print(stats)
```

```python
#  Take the min 

print(np.min(stats))
print(np.min(stats, axis=0)) #  The “Axis=0” command retrieves the row containing the smallest/largest value.
```

```python
#  Take the max
print(np.max(stats))
print(np.max(stats, axis=1)) #  The "Axis=1" command returns the maximum/minimum value of each row.
```

```python
#  Take the mean

np.mean(stats)
```

```python
#  Take the median

np.median(stats)
```

```python
#  Sum all values

print(np.sum(stats))
print(np.sum(stats, axis=0)) #  With the value “Axis=0”, we print the totals of all columns separately.
print(np.sum(stats, axis=1)) #  With the value "Axis=1", we print the totals of all rows separately.
```
